## Allows to type in the textbox

```python
dbutils.widgets.text("tablename","cities","enter the tablename to query")
```

## Allows to choose one from the dropdown

```python
dbutils.widgets.dropdown("widget1", "dev",["dev", "prod", "qa"], "enter the dept name")
```

## Allows to select multiple options from the dropdown

```python
dbutils.widgets.multiselect("widget_2", "dev",["dev", "prod", "qa"],"Enter Dept")
```

## Combobox gives dropdown suggestions and also let the user to type text in the box

```python
dbutils.widgets.combobox("combobox_1", "dev",["dev","prod","qa","admin"],"Select dept")
```
